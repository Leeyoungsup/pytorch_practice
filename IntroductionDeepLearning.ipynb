```python
import numpy as np 
import cv2
import os 
import pandas as pd
import torch
import matplotlib.pyplot as plt
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

선형그래프

```python
t=np.arange(0,5,0.2)
t2=t**2
t3=t**3
mu,sigma=100,15
x=mu+sigma*np.random.randn(10000)
plt.plot(t,t2,label="t2")
plt.plot(t,t3,label="t3")
plt.title("show 2 plots in 1 figure")
plt.legend()
plt.show()
```

서브플롯

```python
plt.subplot(1,2,1)
plt.plot(t,t)
plt.subplot(1,2,2)
plt.plot(t,t3)
plt.show()
```

히스토그램

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title("Histogram of Normal distribution")
plt.hist(x,50,density=1,facecolor='g',alpha=1)
plt.show()
```
